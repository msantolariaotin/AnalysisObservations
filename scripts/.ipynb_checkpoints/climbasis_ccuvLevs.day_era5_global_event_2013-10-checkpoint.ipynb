```python
import sys
sys.path.insert(1, '/home/santolam/MyPythonLibrary/MAR-RECAP/libmar/')
import climbasis as climb
from climbasis import *
import domain as dom
import myplot
import numpy.ma as ma
from myplot import *
import obsinfo as obs
from obsinfo import *
```

```python
Tlevs=xr.open_dataset('/bettik/PROJECTS/pr-regional-climate/santolam/era5/monthly/'+
                      'Tlevs_1000-001_mon.era5_194801_202305_interp.nc')
Ulevs=xr.open_dataset('/bettik/PROJECTS/pr-regional-climate/santolam/era5/monthly/'+
                      'Ulevs_1000-001_mon.era5_194801_202305_interp.nc')
Vlevs=xr.open_dataset('/bettik/PROJECTS/pr-regional-climate/santolam/era5/monthly/'+
                      'Vlevs_1000-001_mon.era5_194801_202305_interp.nc')
```

```python
#Tlevs
```

```python
lev0=850
t=Tlevs.sel(level=lev0)['t']
u=Ulevs.sel(level=lev0)['u']
v=Vlevs.sel(level=lev0)['v']
```

```python
season = 'DJF'

# Select the season
t_season = t.sel(time=t['time'].dt.season == season)
u_season = u.sel(time=u['time'].dt.season == season)
v_season = v.sel(time=v['time'].dt.season == season)

# Compute the climatology (mean over all years for that season)
t_clim = t_season.groupby('time.year').mean('time').mean('year')
u_clim = u_season.groupby('time.year').mean('time').mean('year')
v_clim = v_season.groupby('time.year').mean('time').mean('year')
```

```python
t_clim
```

```python

proj = ccrs.PlateCarree()

# Create figure
fig, ax = plt.subplots(figsize=(16,20), subplot_kw={'projection': proj})

# Add map features
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='lightgray')

# Plot temperature climatology (shading)
c = ax.pcolormesh(
    t_clim.lon, t_clim.lat, t_clim-273.15,
    cmap='coolwarm', shading='auto', transform=ccrs.PlateCarree()
)
cb = plt.colorbar(c, ax=ax, orientation='horizontal', pad=0.05)
cb.set_label('Temperature [K]')
#ax.set_extent([-15, 30, 30, 70], crs=ccrs.PlateCarree())

# Plot wind vectors (thin out for clarity)
skip = 5  # adjust this to reduce arrow density
Q=ax.quiver(
    u_clim.lon[::skip], u_clim.lat[::skip],
    u_clim[::skip, ::skip], v_clim[::skip, ::skip],
    transform=ccrs.PlateCarree(), scale=400
)
ax.quiverkey(
    Q,
    X=0.9, Y=-0.1,  # position in axes coordinates
    U=10,  # reference vector length in m/s
    label='10 m/s',
    labelpos='E')

# Optional: Add gridlines and title
ax.gridlines(draw_labels=True)
ax.set_title(f'Climatological Temperature and Wind Vectors {lev0}hPa (' + season + ')')

plt.show()
```

```python
iyr=1950;fyr=2022
t_period=t_season.sel(time=slice(str(iyr),str(fyr)))
t=t_period.groupby('time.year').mean('time')
par_t=climb.trend_vect(t.year,t,dim="year")

u_period=u_season.sel(time=slice(str(iyr),str(fyr)))
u=u_period.groupby('time.year').mean('time')
par_u=climb.trend_vect(u.year,u,dim="year")

v_period=v_season.sel(time=slice(str(iyr),str(fyr)))
v=v_period.groupby('time.year').mean('time')
par_v=climb.trend_vect(v.year,v,dim="year")
```

```python
par_u[0].max()
```

```python
clevs=np.arange(-0.5,0.52,0.02)
cmap_mod = plt.get_cmap('bwr')
norm = mcolors.BoundaryNorm(boundaries=clevs, ncolors=cmap_mod.N, extend="both")

proj = ccrs.PlateCarree()

# Create figure
fig, ax = plt.subplots(figsize=(16,20), subplot_kw={'projection': proj})

# Add map features
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='lightgray')

# Plot temperature climatology (shading)
c = ax.pcolormesh(
    t_clim.lon, t_clim.lat, 10*par_t[0],
    cmap=cmap_mod,norm=norm, shading='auto', transform=ccrs.PlateCarree()
)
cb = plt.colorbar(c, ax=ax, orientation='horizontal', pad=0.05)
cb.set_label('Temperature [°C/dec]')
#ax.set_extent([-15, 30, 30, 70], crs=ccrs.PlateCarree())

# Plot wind vectors (thin out for clarity)
skip = 4  # adjust this to reduce arrow density
Q=ax.quiver(
    par_u[0].lon[::skip], par_u[0].lat[::skip],
    10*par_u[0][::skip, ::skip], 10*par_v[0][::skip, ::skip],
    transform=ccrs.PlateCarree(),scale=10,  # smaller scale → longer arrows
)
ax.quiverkey(
    Q,
    X=0.9, Y=-0.1,  # position in axes coordinates
    U=0.005,  # reference vector length in m/s
    label='0.005 (m/s)/dec',
    labelpos='E')

# Optional: Add gridlines and title
ax.gridlines(draw_labels=True)
ax.set_title(f'Spatial trend Temperature and Wind Vectors {lev0}hPa ({iyr}-{fyr} {season})')

plt.show()
```
